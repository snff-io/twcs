```python
import webbrowser
import os
from openai import OpenAI
import time

client = OpenAI()

def process_directory(directory):
    """
    Recursively search for description.md files starting from the given directory.
    """
    for root, dirs, files in os.walk(directory):
        if 'urls.md' in files:
            print(f"Skipping directory {root} as urls.md already exists")
            continue
        
        if 'description.md' in files:
            description_path = os.path.join(root, 'description.md')
            with open(description_path, 'r') as f:
                p = f.read().strip()
            
            if not p:
                print(f"Warning: Empty prompt in {description_path}. Skipping...")
                continue
            
            full_prompt = "3d model, cell shaded dramatic lighting narrow field of view " + p,
            full_prompt = " ".join(full_prompt)
            response = client.images.generate(
                model="dall-e-2",
                prompt=full_prompt,
                size="1024x1024",
                quality="standard",
                n=1
            )
            
            image_url = response.data[0].url
            urls_md_path = os.path.join(root, 'urls.md')
            with open(urls_md_path, 'w') as f:
                f.write(image_url)
            print(f"Image URL saved to {urls_md_path}")
  
            webbrowser.get("chromium").open(image_url)
            print("sleep 5")
            time.sleep(5)
            print("sleep 5")
            time.sleep(5)
            print("sleep 5")
            time.sleep(5)

# Start processing from the specified directory
process_directory('./landscapes')
```
