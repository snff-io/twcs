```python
from PIL import Image

def process_image(filename, chunk_size=10):
    img = Image.open(filename)
    width, height = img.size
    
    for y in range(0, height, chunk_size):
        for x in range(0, width, chunk_size):
            chunk = img.crop((x, y, x + chunk_size, y + chunk_size))
            pixels = chunk.getdata()
            r_bg, g_bg, b_bg = 0, 0, 0
            r_fg, g_fg, b_fg = 0, 0, 0
            for i, pixel in enumerate(pixels):
                if i < chunk_size ** 2 / 2:  # Top half for background
                    r_bg += pixel[0]
                    g_bg += pixel[1]
                    b_bg += pixel[2]
                else:  # Bottom half for foreground
                    r_fg += pixel[0]
                    g_fg += pixel[1]
                    b_fg += pixel[2]
            r_bg //= chunk_size ** 2 // 2
            g_bg //= chunk_size ** 2 // 2
            b_bg //= chunk_size ** 2 // 2
            r_fg //= chunk_size ** 2 // 2
            g_fg //= chunk_size ** 2 // 2
            b_fg //= chunk_size ** 2 // 2
            print(f"\033[48;2;{r_bg};{g_bg};{b_bg}m\033[38;2;{r_fg};{g_fg};{b_fg}m▄", end="")
        print()

# Example usage
process_image("/mnt/chromeos/removable/KINGSTON/assets/img/_2ad28d31-22d6-40ef-911e-a421088afb76.jpeg")
```
